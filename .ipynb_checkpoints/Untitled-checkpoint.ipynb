```python
"""
Interactive Grid Program
========================
Place Start, End, and Obstacles on a grid.
  - Select a mode from the left panel
  - Click cells to place items
  - Right-click any cell to erase it
  - Drag in Obstacle / Erase mode to paint
"""

import tkinter as tk
from tkinter import messagebox


# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
#  Constants
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
DEFAULT_ROWS = 15
DEFAULT_COLS = 20
CELL_SIZE    = 38
PADDING      = 2

COLORS = {
    "empty":    "#1e1e2e",
    "empty_h":  "#313244",
    "start":    "#a6e3a1",
    "end":      "#f38ba8",
    "obstacle": "#45475a",
    "grid_bg":  "#11111b",
    "text":     "#cdd6f4",
    "panel_bg": "#181825",
    "btn_bg":   "#313244",
    "btn_act":  "#89b4fa",
    "btn_text": "#cdd6f4",
    "border":   "#6c7086",
}

MODE_INFO = {
    "start":    ("ğŸŸ¢ Start",    COLORS["start"],    "Place the START point"),
    "end":      ("ğŸ”´ End",      COLORS["end"],      "Place the END point"),
    "obstacle": ("â¬› Obstacle", COLORS["obstacle"], "Draw OBSTACLE walls"),
    "clear":    ("âœ– Erase",    COLORS["empty"],    "Erase cells"),
}


# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
#  Main Application
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
class GridApp(tk.Tk):
    def __init__(self):
        super().__init__()
        self.title("Interactive Grid Planner")
        self.configure(bg=COLORS["grid_bg"])
        self.resizable(False, False)

        self.rows    = DEFAULT_ROWS
        self.cols    = DEFAULT_COLS
        self.mode    = tk.StringVar(value="start")
        self.is_drag = False
        self.cells   = {}   # (r, c) -> state str
        self.rects   = {}   # (r, c) -> canvas rect id
        self.start   = None
        self.end     = None

        self._build_ui()
        self._init_grid()
        self._update_status()

    # â”€â”€ UI Layout â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    def _build_ui(self):
        # Title bar
        title_frame = tk.Frame(self, bg=COLORS["panel_bg"], pady=10)
        title_frame.pack(fill="x")
        tk.Label(
            title_frame, text="ğŸ—º  Grid Planner",
            font=("Segoe UI", 18, "bold"),
            bg=COLORS["panel_bg"], fg=COLORS["btn_act"]
        ).pack()
        tk.Label(
            title_frame,
            text="Click to place  â€¢  Right-click to erase  â€¢  Drag to draw obstacles",
            font=("Segoe UI", 9),
            bg=COLORS["panel_bg"], fg=COLORS["border"]
        ).pack()

        # Main body
        body = tk.Frame(self, bg=COLORS["grid_bg"])
        body.pack(padx=14, pady=10)

        self._build_left_panel(body)
        self._build_canvas(body)
        self._build_right_panel(body)

        # Status bar
        self.status_var = tk.StringVar(value="")
        status_bar = tk.Frame(self, bg=COLORS["panel_bg"], pady=6)
        status_bar.pack(fill="x")
        tk.Label(
            status_bar, textvariable=self.status_var,
            font=("Segoe UI", 9), bg=COLORS["panel_bg"], fg=COLORS["border"]
        ).pack()

    # â”€â”€ Left Panel â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    def _build_left_panel(self, parent):
        panel = tk.Frame(parent, bg=COLORS["panel_bg"], padx=12, pady=12)
        panel.pack(side="left", fill="y", padx=(0, 12))

        tk.Label(panel, text="MODE", font=("Segoe UI", 10, "bold"),
                 bg=COLORS["panel_bg"], fg=COLORS["border"]).pack(anchor="w", pady=(0, 8))

        self.mode_buttons = {}
        for key, (label, color, tip) in MODE_INFO.items():
            self._make_mode_btn(panel, key, label, color, tip)

        tk.Frame(panel, bg=COLORS["border"], height=1).pack(fill="x", pady=14)

        tk.Label(panel, text="ACTIONS", font=("Segoe UI", 10, "bold"),
                 bg=COLORS["panel_bg"], fg=COLORS["border"]).pack(anchor="w", pady=(0, 8))

        self._make_action_btn(panel, "ğŸ”„  Clear Grid", self._clear_grid)

    def _make_mode_btn(self, parent, key, label, color, tip):
        frame = tk.Frame(parent, bg=COLORS["panel_bg"])
        frame.pack(fill="x", pady=3)

        tk.Label(frame, text="â—", font=("Segoe UI", 12),
                 bg=COLORS["panel_bg"], fg=color, width=2).pack(side="left")

        btn = tk.Button(
            frame, text=label,
            font=("Segoe UI", 10),
            bg=COLORS["btn_bg"], fg=COLORS["btn_text"],
            activebackground=COLORS["btn_act"], activeforeground="#11111b",
            relief="flat", bd=0, padx=10, pady=6, anchor="w",
            cursor="hand2",
            command=lambda k=key: self._set_mode(k)
        )
        btn.pack(side="left", fill="x", expand=True)

        self.mode_buttons[key] = btn
        self._update_btn_style(key)

    def _make_action_btn(self, parent, text, cmd):
        tk.Button(
            parent, text=text,
            font=("Segoe UI", 10),
            bg=COLORS["btn_bg"], fg=COLORS["btn_text"],
            activebackground="#f5c2e7", activeforeground="#11111b",
            relief="flat", bd=0, padx=10, pady=6, anchor="w",
            cursor="hand2", command=cmd
        ).pack(fill="x", pady=3)

    # â”€â”€ Canvas â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    def _build_canvas(self, parent):
        canvas_w = self.cols * (CELL_SIZE + PADDING) + PADDING
        canvas_h = self.rows * (CELL_SIZE + PADDING) + PADDING

        frame = tk.Frame(parent, bg=COLORS["grid_bg"],
                         highlightbackground=COLORS["border"],
                         highlightthickness=1)
        frame.pack(side="left")

        self.canvas = tk.Canvas(
            frame, width=canvas_w, height=canvas_h,
            bg=COLORS["grid_bg"], highlightthickness=0, cursor="crosshair"
        )
        self.canvas.pack()

        self.canvas.bind("<Button-1>",        self._on_click)
        self.canvas.bind("<B1-Motion>",       self._on_drag)
        self.canvas.bind("<ButtonRelease-1>", lambda e: setattr(self, "is_drag", False))
        self.canvas.bind("<Button-3>",        self._on_right_click)
        self.canvas.bind("<B3-Motion>",       self._on_right_drag)
        self.canvas.bind("<Motion>",          self._on_hover)
        self.canvas.bind("<Leave>",           self._on_leave)

        self._hover_cell = None

    # â”€â”€ Right Panel (legend only) â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    def _build_right_panel(self, parent):
        panel = tk.Frame(parent, bg=COLORS["panel_bg"], padx=12, pady=12)
        panel.pack(side="left", fill="y", padx=(12, 0))

        tk.Label(panel, text="LEGEND", font=("Segoe UI", 10, "bold"),
                 bg=COLORS["panel_bg"], fg=COLORS["border"]).pack(anchor="w", pady=(0, 10))

        for color, name in [
            (COLORS["start"],    "Start Point"),
            (COLORS["end"],      "End Point"),
            (COLORS["obstacle"], "Obstacle"),
            (COLORS["empty"],    "Empty Cell"),
        ]:
            row = tk.Frame(panel, bg=COLORS["panel_bg"])
            row.pack(fill="x", pady=4)
            tk.Canvas(row, width=22, height=22, bg=color,
                      highlightbackground=COLORS["border"],
                      highlightthickness=1).pack(side="left", padx=(0, 8))
            tk.Label(row, text=name, font=("Segoe UI", 9),
                     bg=COLORS["panel_bg"], fg=COLORS["text"]).pack(side="left")

    # â”€â”€ Grid Init â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    def _init_grid(self):
        self.canvas.delete("all")
        self.cells = {}
        self.rects = {}
        self.start = None
        self.end   = None

        for r in range(self.rows):
            for c in range(self.cols):
                x1 = c * (CELL_SIZE + PADDING) + PADDING
                y1 = r * (CELL_SIZE + PADDING) + PADDING
                x2 = x1 + CELL_SIZE
                y2 = y1 + CELL_SIZE
                rid = self.canvas.create_rectangle(
                    x1, y1, x2, y2,
                    fill=COLORS["empty"], outline=""
                )
                self.rects[(r, c)] = rid
                self.cells[(r, c)] = "empty"

    # â”€â”€ Cell Helpers â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    def _get_cell(self, event):
        col = event.x // (CELL_SIZE + PADDING)
        row = event.y // (CELL_SIZE + PADDING)
        if 0 <= row < self.rows and 0 <= col < self.cols:
            return (row, col)
        return None

    def _set_cell(self, cell, state):
        self.cells[cell] = state
        self.canvas.itemconfig(self.rects[cell], fill=COLORS[state])
        self._update_status()

    # â”€â”€ Interactions â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    def _on_click(self, event):
        self.is_drag = True
        cell = self._get_cell(event)
        if cell:
            self._apply_mode(cell)

    def _on_drag(self, event):
        if not self.is_drag:
            return
        cell = self._get_cell(event)
        if cell and self.mode.get() in ("obstacle", "clear"):
            self._apply_mode(cell)

    def _on_right_click(self, event):
        cell = self._get_cell(event)
        if cell:
            self._erase_cell(cell)

    def _on_right_drag(self, event):
        cell = self._get_cell(event)
        if cell:
            self._erase_cell(cell)

    def _on_hover(self, event):
        cell = self._get_cell(event)
        if cell == self._hover_cell:
            return
        if self._hover_cell and self.cells.get(self._hover_cell) == "empty":
            self.canvas.itemconfig(self.rects[self._hover_cell], fill=COLORS["empty"])
        self._hover_cell = cell
        if cell and self.cells.get(cell) == "empty":
            self.canvas.itemconfig(self.rects[cell], fill=COLORS["empty_h"])

    def _on_leave(self, event):
        if self._hover_cell and self.cells.get(self._hover_cell) == "empty":
            self.canvas.itemconfig(self.rects[self._hover_cell], fill=COLORS["empty"])
        self._hover_cell = None

    def _apply_mode(self, cell):
        mode = self.mode.get()
        if mode == "start":
            if self.start and self.start != cell:
                self._set_cell(self.start, "empty")
            self.start = cell
            self._set_cell(cell, "start")
        elif mode == "end":
            if self.end and self.end != cell:
                self._set_cell(self.end, "empty")
            self.end = cell
            self._set_cell(cell, "end")
        elif mode == "obstacle":
            if self.cells[cell] not in ("start", "end"):
                self._set_cell(cell, "obstacle")
        elif mode == "clear":
            self._erase_cell(cell)

    def _erase_cell(self, cell):
        state = self.cells.get(cell, "empty")
        if state == "start":
            self.start = None
        elif state == "end":
            self.end = None
        self._set_cell(cell, "empty")

    # â”€â”€ Mode Selection â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    def _set_mode(self, key):
        self.mode.set(key)
        for k in self.mode_buttons:
            self._update_btn_style(k)
        self._update_status()

    def _update_btn_style(self, key):
        btn = self.mode_buttons.get(key)
        if not btn:
            return
        active = (self.mode.get() == key)
        btn.configure(
            bg=COLORS["btn_act"] if active else COLORS["btn_bg"],
            fg="#11111b"          if active else COLORS["btn_text"]
        )

    # â”€â”€ Actions â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    def _clear_grid(self):
        if messagebox.askyesno("Clear Grid", "Clear all cells?"):
            self._init_grid()
            self._update_status()

    # â”€â”€ Status Bar â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    def _update_status(self):
        mode = self.mode.get()
        tip_map = {
            "start":    "Click a cell to place the START point",
            "end":      "Click a cell to place the END point",
            "obstacle": "Click or drag to draw OBSTACLES",
            "clear":    "Click or drag to ERASE cells",
        }
        start_info = f"âœ… Start @ {self.start}" if self.start else "âŒ No start"
        end_info   = f"âœ… End @ {self.end}"     if self.end   else "âŒ No end"
        self.status_var.set(
            f"Mode: {mode.upper()}  â€¢  {tip_map[mode]}  |  {start_info}  â€¢  {end_info}"
        )


# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
#  Entry Point
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
if __name__ == "__main__":
    app = GridApp()
    app.mainloop()
```
